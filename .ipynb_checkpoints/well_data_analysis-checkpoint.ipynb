##### *Import Libraries*

```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
matplotlib.rcParams['figure.figsize'] = (20,10)
```

##### *Load Data Set*

```python
well_data = pd.read_csv('./796_wells_lat_long.csv')
well_data.head()
```

```python
# basic statistical description of the data
well_data.describe()
```

##### *Clean Data Set*

```python
# setting column names for easy referencing
convenient_column_names = ['R_date','Lat','Lon','1995','2000','2005','2010','2015']
well_data.columns = convenient_column_names
well_data.describe()
```

```python
# finding out the max and minimum well water level values of the dataset
max_elevation = max(well_data.iloc[:,3:].max())
min_elevation = min(well_data.iloc[:,3:].min())
```

```python
def classify(data,column,n):
    """
    adds a column to dataset by classifing based on a column.
    ---
    parameters
    data -- DataFrame -- dataframe
    column -- str -- column name
    n -- int -- number of segments. ie, n-1 ranges
    returns a DataFrame
    """
    range_points = np.linspace(min_elevation,max_elevation,n)
    criteria = [data[column].between(range_points[key-1],range_points[key]) for key,value in list(enumerate(range_points))[1:]]
    values = [f'{range_points[key-1]:.2f} ~ {range_points[key]:.2f}' for key,value in list(enumerate(range_points))[1:]]
    data['Well GWL Range'] = np.select(criteria, values, 0)
    
    return data

# making a function that returns a data set containing valid ie not null values of a certain year
def clean_data(column,df=well_data,n=6):
    columns_to_keep = ['Lat','Lon',column]
    out_df = df.copy().loc[:,columns_to_keep]
    out_df = out_df[out_df[column].notna()]
    out_df = gpd.GeoDataFrame(out_df, geometry=gpd.points_from_xy(out_df['Lon'], out_df['Lat']))
    out_df = classify(out_df,column,n)
    return out_df
    
```

```python
# as an example, this is cleaned 1995 data
clean_data('1995')
```

```python
# storing cleaned data in a dictionary named df for easy future use
well_all = dict((x,clean_data(x)) for x in convenient_column_names[3:])
well_all
```

### Load Bangladesh Shape file

```python
# Plot map of Bangladesh
bd_map_ax = gpd.read_file('./Bangladesh.shp').plot(edgecolor='black', color='white')
```

```python
# this function plots the cleaned dataset on the map of Bangladesh
def generate_plot(column_name,geometry_col='geometry',df=well_all):
    country_ax = gpd.read_file('./Bangladesh.shp').plot(edgecolor='black', color='white')
    da = df[column_name].copy()
    plt.figure(figsize=(5,9))
    country_ax = gpd.read_file('./Bangladesh.shp').plot(edgecolor='black', color='white')
    
    ax = sns.scatterplot(data=da,x='Lon',y='Lat',hue='Well GWL Range', ax=country_ax, marker='+')
    ax.set_title(column_name);
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')
    return ax
```

```python
# Now plotting the five year cleaned data separately on map of Bangladesh
# Also saving an image of each plot 
# the figures are also saved in the dictionary `figures`
figures = {}
for x in convenient_column_names[3:]:
    ax = generate_plot(x)
    fig = ax.get_figure()
    fig.savefig(x+'.png')
    figures[x] = ax
figures
```
